```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# data dummy
x = np.random.RandomState(5).rand(50) * 10
```

```python
x
```

```python
y = 2 * x - 5 + np.random.RandomState(1).randn(50)
```

```python
y
```

```python
sns.set_style('whitegrid')
sns.regplot(x=x,y=y)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x.reshape(-1,1),y)
```

```python
y = np.sin(x)
```

```python
sns.regplot(x=x,y=y,order=7,ci=None,truncate=True)
```

```python
# polynomial, make_pipline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
model = make_pipeline(PolynomialFeatures(8),LinearRegression())
```

```python
model.fit(x.reshape(-1,1),y)
```
